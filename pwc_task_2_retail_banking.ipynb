Preprocessing :-

```python
from sklearn.model_selection import train_test_split
X = retail_banking.drop('y',axis=1)
y=retail_banking['y']
```

```python
X_train,X_val,y_train,y_val = train_test_split(X,y,train_size=0.8)
```

```python
one_hot_columns = ['job','marital','default','housing','loan','contact','poutcome']
ordinal_columns = ['education','month','day_of_week']

education_order = ['unknown', 'illiterate','basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'professional.course', 'university.degree']
month_order = ['jan','feb','mar', 'apr', 'may', 'jun','jul', 'aug','sep','oct','nov','dec']
day_order = ['mon','tue','wed','thu','fri']
```

```python
one_hot_df_train = pd.get_dummies(X_train[one_hot_columns], columns=one_hot_columns,drop_first=True)

# Ordinal encoding
ordinal_df_train = X_train[ordinal_columns].replace({
    'education': {val: idx for idx, val in enumerate(education_order)},
    'month': {val: idx for idx, val in enumerate(month_order, start=1)},  # Starting from 1 to match your desired output
    'day_of_week': {val: idx for idx, val in enumerate(day_order, start=1)}
})

X_train_encoded = pd.concat([one_hot_df_train, ordinal_df_train], axis=1)


one_hot_df_val = pd.get_dummies(X_val[one_hot_columns], columns=one_hot_columns,drop_first=True)

# Ordinal encoding
ordinal_df_val = X_val[ordinal_columns].replace({
    'education': {val: idx for idx, val in enumerate(education_order)},
    'month': {val: idx for idx, val in enumerate(month_order, start=1)},  # Starting from 1 to match your desired output
    'day_of_week': {val: idx for idx, val in enumerate(day_order, start=1)}
})

X_val_encoded = pd.concat([one_hot_df_val, ordinal_df_val], axis=1)
X_val_encoded

X_train_encoded,X_val_encoded = X_train_encoded.align(X_val_encoded,join='left',axis=1,fill_value=0)
```

```python
X_val_encoded
```

# Model building

Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, f1_score

model = LogisticRegression(max_iter=500,random_state=0)
model.fit(X_train_encoded,y_train)
pred_y = model.predict(X_val_encoded)

score = f1_score(y_val,pred_y,pos_label='yes')
score
```

```python
classification_report(y_val,pred_y)
```

```python
sns.heatmap(confusion_matrix(y_val,pred_y),annot=True,fmt='d',cmap='Blues',cbar=False)
```

# The f1_score and the confusion matrix shows that the data is imbalanced. For now i have ignored the performance of the logistic regression model. In order to increase the model performance we have to perform operations like oversampling, undersampling etc. 

Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=500)
model_rf.fit(X_train_encoded,y_train)
y_pred_rf = model.predict(X_val_encoded)

score_rf = f1_score(y_val,y_pred_rf,pos_label='yes')
score_rf
```

The result of both Logistic Regression model and Random forest classifier are same. For this dataset both the models are performing quite same.

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

ros = RandomOverSampler()
X_train_ros,y_train_ros = ros.fit_resample(X_train_encoded,y_train)
rus = RandomUnderSampler()
X_train_rus,y_train_rus = rus.fit_resample(X_train_ros,y_train_ros)
final_model = LogisticRegression(max_iter=1000)
final_model.fit(X_train_rus,y_train_rus)
y_pred_final = final_model.predict(X_val_encoded)
f1_score(y_pred_final,y_val ,pos_label='yes')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer

param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'lbfgs', 'saga']
}

f1_scorer = make_scorer(f1_score, pos_label='yes')
grid_search = GridSearchCV(final_model, param_grid, scoring=f1_scorer, cv=5)
grid_search.fit(X_train_rus, y_train_rus)

best_model = grid_search.best_estimator_
```
